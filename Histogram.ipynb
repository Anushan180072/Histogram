```python
import numpy as np
import matplotlib.pyplot as plt
```

**Histogram for 1DArray**

```python
arr=[15,22,20,30,80,24,68,25,77,90]
b=[10,21,31,41,51,61,71,81,91,101]
plt.hist(arr,bins=b,rwidth=0.5,facecolor="green")  #edgecolor="red"
plt.xticks(b)
plt.xlabel("age")
plt.ylabel("count")
plt.title("Age Count")
plt.show()
```
